```python
import numpy as np

# Matplotlib einrichten
import matplotlib.pyplot as plt
%matplotlib inline

from astropy.io import fits
```

```python
from astropy.utils.data import download_file
image_file = download_file('http://data.astropy.org/tutorials/FITS-images/HorseHead.fits', cache=True )
```

# Öffnen von FITS-Dateien und Laden der Bilddaten

```python
hdu_list = fits.open(image_file)
hdu_list.info()
```

```python
image_data = hdu_list[0].data
```

```python
print(type(image_data))
print(image_data.shape)
```

```python
hdu_list.close()
```

```python
image_data = fits.getdata(image_file)
```

```python
print(type(image_data)) # Zeige den Python-Typ für image_data
print(image_data.shape) # Zeige die Anzahl der Pixel pro Seite im 2D-Bild an
```

```python
plt.imshow(image_data, cmap='gray')
plt.colorbar()

# Um mehr Farbkarten zu sehen
# http://wiki.scipy.org/Cookbook/Matplotlib/Show_colormaps
```
